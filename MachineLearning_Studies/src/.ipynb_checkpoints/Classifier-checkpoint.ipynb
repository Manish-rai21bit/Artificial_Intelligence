```python
#Reading the training dataset
import numpy as np
import pandas as pd
import os
os.chdir('/Users/manishrai/Desktop/DSMLAI/')
```

```python
## Defining function to read the csv file

def get_read_file_func(filename):
    data = np.genfromtxt(filename, delimiter=',', skip_header=1)
    data_col = np.genfromtxt(filename, delimiter=',', max_rows = 1, dtype= None)
    print("Successfully Loaded", filename)
    return data, data_col
```

```python
filename = '/Users/manishrai/Desktop/DSMLAI/MNIST/Data/train.csv'
train_data, train_data_col = get_read_file_func(filename)
```

```python
# Viewing the loaded file in Pandas dataframe
df1 = pd.DataFrame(data=train_data, columns=train_data_col)  # 1st row as the column names
df1[:10]
```
