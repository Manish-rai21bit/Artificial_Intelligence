### Probability Theory
The key concept in pattern recognition is the uncertainity that arises due to noise aswell as due to the finite size of the data. Probability theory forms one of the central piece in patter recognition.
- Product Rule and Sum Rule(Marginal Probability) of probability can be pictured as a box with gridlines and the grids in the x-axis and y-axis corresponding to random variables X and Y
- The probability that a Random Variable \begin{equation} \ X \text{will take a value} \  x_i \text{and Random Variable }\ Y \text{will take a value } \ y_j \text{is written as }\end{equation} \begin{equation} \ p(X = x_i, Y = y_j) \end{equation} and is called the _Joint Probability._
    - \begin{equation} \ p(X = x_i, Y = y_j) = n_ij/N \text{     ------------1} \end{equation}
- Sum Rule says that the probability of X taking the value x_i irrespective of the valye of Y is : <br>
\begin{equation} \ p(X=x_i) = \Sigma\ (p(X = x_i, Y = y_j)) = c_i/N\text{     Summation over all j's ------------2} \end{equation} 
\begin{equation} \ p(X=x_i) = \Sigma\ (p(X = x_i, Y = y_j))\text{     -----This is called the Sum Rule, Summation over all j's} \end{equation}  
<br>
- Conditional Probability - if we consider only those instances where random variable \begin{equation} \ X = x_i\text{ , then the fraction of such instances for which} \ Y = y_j\text {is written as:} \end{equation} \begin{equation} \ p(Y = y_i | X = x_i)\end{equation}  and is called conditional probaility.<br>

\begin{equation} \ p(Y = y_i | X = x_i) = n_ij / c_i \text{     ------------3} \end{equation} 

- using 1, 2 & 3:<br>
\begin{equation} \ p(X = x_i, Y = y_j) = n_ij/N = (n_ij/c_i) * (c_i/N) = p(Y = y_i | X = x_i) * p(X=x_i) \text{----- This is the product rule} \end{equation}  

```python
#visualize the above formulas in this grid
plt.figure(figsize=(4,4))
plt.grid()
plt.xlabel('x_i')
plt.ylabel('y_j')
plt.show()
```

\begin{equation} \ p(Y|X) = p(X|Y) * p(Y) / p(X)\end{equation}

This equates to - _posterior  Likelihood * prior_ <br>

##### The Gaussian distribution

For the case of a real valued variable x, the gaussian distribution is defined as :

\begin{equation}\ N(x |\mu , \sigma ^2) = \frac{1}{\sqrt{(2  \pi  \sigma ^2)}}  exp(\frac{-1}{2 \sigma^2}(x - \mu)^2)\end{equation}

And the area under the gaussian curve is normalised:
\begin{equation}\int  N(x |\mu , \sigma ^2)dx = 1\end{equation}

And, expectation of a function x under gaussion distribution is given be :
\begin{equation}\ E[x] = \int  N(x |\mu , \sigma ^2)xdx = \mu\end{equation}

The second order moment is given by :
\begin{equation}\ E[x^2] = \int  N(x |\mu , \sigma ^2)x^2dx = \mu^2 + \sigma^2\end{equation}

Using the above 2 equations, the variance of the function x is given by :
\begin{equation}\ var[x] = E[x^2] - E[x]^2 = \sigma^2\end{equation}

_Similarly, we can define the Gaussian distribution of a D-dimensional vector x in continuous variable_

```python
# Defining a Gaussian Distribution:
def get_norm_dist_func(x, mu, sigma):
    return (1/math.sqrt(2*math.pi*sigma*sigma))*math.exp((-1/(2*sigma*sigma))* (x-mu)*(x-mu))
```

```python
random.seed(1)
x = sorted(list(random.uniform(-5,5) for i in range(100)))
y = list(get_norm_dist_func(x[i], 0, 1) for i in range(100))

plt.plot(x, y)
plt.xlim(-5, 5)
plt.ylim(0, 0.5)
plt.title('Gaussian Curve')
plt.xlabel('x')
plt.ylabel('normal value')
plt.vlines(0, 0, N, linewidth=2, alpha=0.2)
plt.vlines(-1, 0, N, linewidth=2, alpha=0.2)
plt.vlines(1, 0, N, linewidth=2, alpha=0.2)
plt.vlines(-2, 0, N, linewidth=2, alpha=0.2)
plt.vlines(2, 0, N, linewidth=2, alpha=0.2)
plt.vlines(-3, 0, N, linewidth=2, alpha=0.2)
plt.vlines(3, 0, N, linewidth=2, alpha=0.2)
plt.show()
```

**Maximum likelihood estimation:** A maximum likelihood approach does not attempt to formulate models for the parameter priors, $p(\mathbf{w})$, or the parameter posterior probabilities, $p(\mathbf{w}|D)$. Rather it views the data, $D$, as fixed and attempts to determine the model parameters, $\mathbf{w}$, by maximizing the likelihood function. As we will frequently use this approach, it is useful to understand the likelihood function.

The likelihood function is simply the product of the probability of all the individual data points, $d_i \in D$, under the probability model, $p_{\mathbf{w}}$. Note that this definition implicitly assumes these data points are independent events.

$L(\mathbf{w}; D) = \prod_{i=1}^N p_{\mathbf{w}}(d_i)$ 

Out of mathematical convenience, we will most often work with the log-likelihood function (which turns the product into a sum by properties of the log function), i.e. the logarithm of $L(\mathbf{w}; D)$, defined as 

$l(\mathbf{w};D) = \sum_{i=1}^N l(\mathbf{w};d_i) = \sum_{i=1}^N \log p_{\mathbf{w}}(d_i)$ 

where we recall that $\log(ab) = \log(a) + \log(b)$.

The method of maximum likelihood chooses the value $\mathbf{w} = \widehat{\mathbf{w}}$ that maximizes the log-likelihood function. We will also often work with an error function, $E(\mathbf{w})$, defined as the negative of the log-likelihood function 

$E(\mathbf{w}) = -l(\mathbf{w};D)$ 

where we note $-\log(a) = \log(1/a)$

#### Model Selection:
The performance of the training set is not a good indicator of the goodness of the model, as the model might likely be facing overfitting. For such case, if the data is plentiful we can separete the data into:
    - Training set
    - Validation set
    - Test Set
where we select the best model out of Training and Validation set. <br>

If the data is not plentiful for the validation set, we can use the cross-validation. This allows a proportion of _(S-1)/S_ of the available data to be used for training while making use of the whole data to access the performance. <br>
When the data is particularly scarce it may be appropriate to consider _S = N_. In which case it is called _leave one out technique_. <br>

We need a measure of performance which depends only on the training data and which does not suffer from bias. :
    - Akaike information criteria (AIC)

### The curse of Dimensionality

How does the volume of a sphere that lies between radius r = 1 - e and r = 1 varies?
<br>
The volume of a sphere in higher dimension can be written as:
\begin{equation}\ V_D(r) = K_D r^D\end{equation}
Thus the required fraction is given by:

\begin{equation}\frac{(V_D(1) - V_D(1 - e)}{V_D(1)} = 1 - (1-e)^D\end{equation}

As the number of dimensions, in Machine Learning, increases the amount of data needed to genetare accuracy grows exponentially.<br>
Plotting this to see how the fraction changes!!

```python
e = sorted(list(random.uniform(0, 1) for i in range(50)))

def get_volume_frac_func(D, e):
    return list(1 - (1 - e[i])**D for i in range(len(e)))
```

```python
D1 = get_volume_frac_func(1, e)
D2 = get_volume_frac_func(2, e)
D3 = get_volume_frac_func(3, e)
D4 = get_volume_frac_func(4, e)
D5 = get_volume_frac_func(5, e)
D15 = get_volume_frac_func(15, e)
```

```python
plt.plot(e, D1)
plt.plot(e, D2)
plt.plot(e, D3)
plt.plot(e, D4)
plt.plot(e, D5)
plt.plot(e, D15)
plt.xlabel('e')
plt.ylabel('volume fraction')
plt.show()
```

We see that the fraction of volume between the thin crust increases as we go higher up the dimension.

### Decision Theory
Determining p(x, t) from a set of training data is an examplke of inference. <br>
Decision theory tell us how to make optimal decision geiven the appropriate probabilities. We'll see that the decision stage is generally very simple once we have solved the inference problem. <br>
**Example:** A medical diagnosis problem in which we have to determine if it's a cancer or not based on the X-ray image. 
The X-ray is a set of pixed values which can be written as an input vector **x** with target variable **t** hvaing 2 classes C1/0 (no cancer) & C2/1 (cancer). <br>
Taking a bayesian approach. Given an image x, our goal is to decide which of the classes to assign the image to. We are interested in probabilities of 2 classes, given by p(C_k| x). Using bayes probability this is given by:

\begin{equation}\ p(C_k| x) = \frac{p(x| C_k)p(C_k)}{p(x)}\end{equation}

\begin{equation}\ p(C_k) \text{ - is the prior probability for class} C_k\end{equation}
This represents the probability the the patient has cancer before we take an X-ray. 

\begin{equation}\ P(C_k|x) \text{ - is the posterion probability} \end{equation}
This is the revised probability in light of the X-ray image. <br>

If out aim is to minimize the chances of wrongly classifying the X-ray, we'd choose the class having the higher posterior probability.

_This same theory is used for classification of mails into spam vs non-spam_

**Minimizing the misclassification rate** <br>
\begin{equation}\ p(mistake) = p(x \in R_1, C_2) + p(x \in R_2, C_1) = \int^1{p(x, C_2)dx} + \int^2{p(x, C_1)dx} \end{equation} 
1 - region1; 2 - region2<br>
For a more general case of K classes:
\begin{equation}\ p(correct) = \Sigma^K_1\text{ }p(x \in R_K, C_K) = \Sigma^K_1\text{ } \int{p(x, C_K)dx} \end{equation} 

**Minimizing the expected loss:** - In the case of cancer treatment problem, the loss incurred in case the patient in classified as normal but has cancer will be fatal and will lead to preature death. The loss in this case is more severe and the idea is to give more waitage to this this this case in the Loss matrix. Eventually we minimize the loss function. <br> 
Loss Matric for cancer treatment =

 \end{matrix}

### Information Theory
Considering a random variable x, we ask how much information is recieved when we observe a specefic value of this variable. The amound of information can be viewed a sthe 'degree of surprise' on learning the value of x. <br>
- Entropy - The average amount of information needed to specify the state of a random variable. And, it's given as:

\begin{equation}\ H[x] = -\Sigma_x p(x)\log_2 p(x) \end{equation}
